```python
import nltk
from nltk.corpus import state_union
#from nltk.tokenize import PunktSentenceTokenizer
import re
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
train_text = state_union.raw("D:\\Natural language processing\\Master_Thesis_2022\\sample.txt")
```

```python
sentences = nltk.sent_tokenize(train_text)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words =nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words ('english'))]
    sentences[i] = ' '.join(words)
```

```python
corpus = []
for i in range(len(sentences)):
    clean_word = re.sub('[^a-zA-Z]', ' ', sentences[i])
    clean_word = clean_word.lower()
    clean_word = clean_word.split()
    clean_word = [lemmatizer.lemmatize(word) for word in clean_word if not word in set(stopwords.words('english'))]
    clean_word = ' '.join(clean_word)
    corpus.append(clean_word)
```

```python
def process_content():
    try:
        for i in corpus:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)
    except Exception as e:
        print(str(e))
        
```
